```python
import pandas as pd
```

```python
wc_teams = ['Russia', 'Saudi Arabia', 'Egypt', 'Uruguay', 'Portugal', 'Spain',
       'Morocco', 'Iran', 'France', 'Australia', 'Peru', 'Denmark',
       'Argentina', 'Iceland', 'Croatia', 'Nigeria', 'Brazil',
       'Switzerland', 'Costa Rica', 'Serbia', 'Germany', 'Mexico',
       'Sweden', 'Korea', 'Belgium', 'Panama', 'Tunisia', 'England',
       'Poland', 'Senegal', 'Columbia', 'Japan']
```

```python
players = pd.read_csv("CompleteDataset.csv", low_memory=False)
```

```python
players.drop(['Unnamed: 0','Photo', 'Flag', 'Club Logo'], axis=1, inplace=True)
```

```python
# figure out!!
#players.apply(pd.to_numeric, errors='ignore')
```

```python
# only need players that are in WC
wc_players = players[players.Nationality.isin(wc_teams)]
```

```python
wc_players.head()
```

```python
# need the mean of other stats
teams = wc_players.groupby('Nationality')['Overall', 'Age'].mean()
```

```python
teams.index.names = ['Team']
```

```python
teams.sort_values('Overall', ascending=False)
```

```python
wc2018 = pd.read_csv('World Cup 2018_Dataset.csv')
wc2018.head()
```

```python
teams_wc_hist = wc2018[['Team','Previous \nappearances', 'Previous \ntitles', 'Previous\n finals', 
                        'Previous\n semifinals', 'Current \nFIFA rank']]
```

```python
teams_wc_hist.dropna()
```

```python
teams_wc_hist.replace({'Porugal': 'Portugal', 'IRAN': 'Iran', 'Costarica': 'Costa Rica'}, inplace=True)
```

```python
teams_wc_hist.set_index('Team')
```

```python
teams_fin = teams.join(teams_wc_hist.set_index('Team'))
```

```python
teams_fin.rename(columns={'Overall':'Squad strength'}, inplace=True)
```
